## Specify the forecasting pipeline.

The pipeline should include all steps of preprocessing, statistical modeling, machine learning, and hyperparameter optimization.

```python
https://github.com/fraunhoferportugal/tsfel
https://github.com/blue-yonder/tsfresh
```

```python
from timegym import Pipeline
from sklearn.decomposition import PCA
from xgboost import XGBRegressor

pipeline = Pipeline()
pipeline.add(PCA(0.95))
pipeline.add(XGBRegressor)
```

## Test the pipeline on synthetic data

Let's test this pipeline for the most common theoretical problems found in forecasting

After a few minutes (depends on the pipeline) you should see the results:

```python
pipeline.plot_tests();
```

```python
pipeline.plot_error();
```
